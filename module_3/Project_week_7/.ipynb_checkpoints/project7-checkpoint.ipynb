```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression

# Initialize models from each group
models = {
    'RidgeClassifier': make_pipeline(StandardScaler(), RidgeClassifier(random_state=42)),
    'SVC': make_pipeline(StandardScaler(), SVC(random_state=42)),
    'RandomForestClassifier': RandomForestClassifier(random_state=42),
    'GradientBoostingClassifier': GradientBoostingClassifier(random_state=42),
    'LogisticRegression': make_pipeline(StandardScaler(), LogisticRegression(random_state=42))
}
```

```python
from sklearn.metrics import classification_report

# Dictionary to store the classification reports for each model
classification_reports = {}

# Fit models and print classification report for each
for name, model in models.items():
    # Fit the model on the training data
    model.fit(X_train_smote, y_train_smote)
    
    # Predict on the test data
    y_pred = model.predict(X_test)
    
    # Store the classification report
    classification_reports[name] = classification_report(y_test, y_pred, target_names=['Non-Bankrupt', 'Bankrupt'])
    
    # Print the classification report
    print(f"Model: {name}")
    print(classification_reports[name])
```

```python
# Update the Logistic Regression model with a higher number of iterations
models['LogisticRegression'] = make_pipeline(
    StandardScaler(),
    LogisticRegression(random_state=42, max_iter=1000)  # Increased from the default 100 to 1000
)
```

```python
# Fit the updated Logistic Regression model
model = models['LogisticRegression']
model.fit(X_train_smote, y_train_smote)

# Predict on the test data
y_pred = model.predict(X_test)

# Print the updated classification report for Logistic Regression
print("Updated Model: LogisticRegression")
print(classification_report(y_test, y_pred, target_names=['Non-Bankrupt', 'Bankrupt']))
```

```python
from sklearn.model_selection import GridSearchCV

# Create a parameter grid to test
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Create a base model
rf = RandomForestClassifier(random_state=42)

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train_smote, y_train_smote)

# Find the best hyperparameters
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")

# Instantiate the best model based on the grid search
best_rf = RandomForestClassifier(**best_params, random_state=42)

# Fit the best model
best_rf.fit(X_train_smote, y_train_smote)

# Predict on the test data
y_pred_best_rf = best_rf.predict(X_test)

# Print the classification report for the best Random Forest model
print("Best Random Forest Classifier")
print(classification_report(y_test, y_pred_best_rf, target_names=['Non-Bankrupt', 'Bankrupt']))
```
