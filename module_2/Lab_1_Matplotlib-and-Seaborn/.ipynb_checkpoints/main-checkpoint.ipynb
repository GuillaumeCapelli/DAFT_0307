#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
plt.figure()

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b-', linewidth=2, label='y = 2x')
plt.title('y = 2x', fontweight='bold', fontsize=12)
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 2, 2)
plt.plot(x, z, 'r--', linewidth=2, label='y = x^2')
plt.title('y = x^2', fontweight='bold', fontsize=12)
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, y, 'b-', linewidth=2, label='y = 2x')
ax[1].plot(x, z, 'r--', linewidth=2, label='y = x^2')

ax[0].set_title('y = 2x', fontweight='bold', fontsize=12)
ax[1].set_title('y = x^2', fontweight='bold', fontsize=12)

ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# Data
x = np.arange(0,100)
y1 = x**2
y2 = np.exp(x)

# Create the first subplot with a normal scale
plt.figure(figsize=(12, 6))

# First subplot: Normal scale
ax1 = plt.subplot(1, 2, 1)
ax1.plot(x, y1, 'b-', linewidth=2, label='y = x^2')
ax1.plot(x, y2, 'r--', linewidth=2, label='y = e^x')
ax1.set_title('Normal Scale', fontweight='bold', fontsize=12)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Second subplot: Logarithmic scale
ax2 = plt.subplot(1, 2, 2)
ax2.plot(x, y1, 'b-', linewidth=2, label='y = x^2')
ax2.plot(x, y2, 'r--', linewidth=2, label='y = e^x')
ax2.set_yscale('log')
ax2.set_title('Logarithmic Scale', fontweight='bold', fontsize=12)
ax2.set_xlabel('x')
ax2.set_ylabel('y (log scale)')
ax2.legend()

plt.tight_layout()
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fig, ax = plt.subplots(nrows=2, figsize=(12, 4))
```

```python
import os

# Create a directory named 'figures'
if not os.path.exists('figures'):
    os.makedirs('figures')

# MEDIAN STEPS BY MONTH_ENCODED
median_steps_month = fitbit.groupby('Months_encoded')['Steps'].median()
plt.sca(ax[0])  # setting the current axis to ax[0]
median_steps_month.plot(kind='line')
plt.title("Median Steps by Month")
plt.xticks(list(range(1, 13)), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# MEDIAN STEPS BY WORK_OR_WEEKEND
median_steps_work_or_weekend = fitbit.groupby('Work_or_Weekend')['Steps'].median()
plt.sca(ax[1])  # setting the current axis to ax[1]
median_steps_work_or_weekend.plot(kind='bar', alpha=0.75)
plt.title("Median Steps: Workdays vs. Weekend")
plt.xticks([0, 1], ['Weekend', 'Workdays'], rotation=0)

plt.tight_layout()
plt.savefig('figures/median_steps_plots.png')  # Saving the figures
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for feature in features:
    plt.figure(figsize=(6,4))
    plt.scatter(fitbit[feature], fitbit['Steps'], alpha=0.5)
    plt.title(f"{feature} vs. Steps")
    plt.xlabel(feature)
    plt.ylabel('Steps')
    plt.tight_layout()
    plt.savefig(f'figures/{feature}_vs_steps.png')  # Saving the figures
    plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5))

bins_values = [10, 20, 50]
for ax, bins in zip(axes, bins_values):
    ax.hist(titanic['age'].dropna(), bins=bins, edgecolor='black')
    ax.set_title(f'Bins: {bins}')
    ax.set_xlabel('Age')
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```
