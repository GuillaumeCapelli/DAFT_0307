# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data_url = 'https://raw.githubusercontent.com/ta-data-par/DAFT_0307/main/module_2/Lab_2_%20Plotting-Multiple-Data-Series/liquor_store_sales.csv'
data = pd.read_csv(data_url)
data
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
grouped = data.groupby('ItemType').sum()[['RetailSales', 'RetailTransfers', 'WarehouseSales']]
grouped.plot(kind='bar', figsize=(10, 6))
plt.title("Total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type")
plt.ylabel("Total")
plt.show()
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
top_suppliers = data.groupby('Supplier').sum()[['RetailSales', 'RetailTransfers', 'WarehouseSales']]
top_suppliers['Total Sales'] = top_suppliers.sum(axis=1)
top_10_suppliers = top_suppliers.sort_values('Total Sales', ascending=False).head(10)
top_10_suppliers['Total Sales'].plot(kind='barh', figsize=(10, 6))
plt.title("Top 10 Suppliers with the Most Total Sales")
plt.xlabel("Total Sales")
plt.show()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
data['Date'] = pd.to_datetime(data['Year'].astype(str) + '-' + data['Month'].astype(str) + '-01')
monthly_sales = data.groupby('Date').mean()[['RetailSales', 'RetailTransfers', 'WarehouseSales']]
monthly_sales.plot(figsize=(10, 6))
plt.title("Average Retail Sales, Retail Transfers, and Warehouse Sales Over Time")
plt.ylabel("Average")
plt.show()
```

## 4. Plot the same information as above but as a bar chart.

```python
monthly_sales.plot(kind='bar', figsize=(12, 7))
plt.title("Average Retail Sales, Retail Transfers, and Warehouse Sales Over Time")
plt.ylabel("Average")
plt.show()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
item_monthly_sales = data.groupby(['Date', 'ItemType']).sum()['RetailSales'].unstack()
item_monthly_sales.plot(figsize=(10, 6))
plt.title("Retail Sales by Item Type Over Time")
plt.ylabel("Total Sales")
plt.show()
```

## 6. Plot the same information as above but as a bar chart.

```python
item_monthly_sales.plot(kind='bar', figsize=(12, 7), stacked=True)
plt.title("Retail Sales by Item Type Over Time")
plt.ylabel("Total Sales")
plt.show()
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
sns.lmplot(x="RetailSales", y="RetailTransfers", data=data, hue="ItemType", fit_reg=False, legend=True)
plt.title("Relationship between Retail Sales and Retail Transfers")
plt.show()
```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.

*Hint: Seaborn's pairplot may be your best option here.*

```python
sns.pairplot(data[['RetailSales', 'RetailTransfers', 'WarehouseSales', 'ItemType']], hue='ItemType', diag_kind="kde", kind="scatter", palette="husl")
plt.show()
```
