```python
import pandas as pd
```

```python
PATH = r"C:\Users\PC\Ironhack\Course\Week4\Project\ds_salaries.csv"
# https://www.kaggle.com/datasets/arnabchaki/data-science-salaries-2023
# https://www.kaggle.com/datasets/ruchi798/data-science-job-salaries

countries_dict = {"FR":"France", 
                  "US":"United States",
                  "CA":"Canada",
                  "ES":"Spain",
                  "IN":"India",
                  "DE":"Germany",
                  "GB":"United Kingdom"
}
```

```python
def myval_count(data, n=30):
    # Display all the modalities for each column of a df
    for col in data.columns:
        print('\n', data[col].value_counts().head(n))
        
```

```python
# Load and explore the data
data = pd.read_csv(PATH)

display(data)
print(data.describe())
data.info()
```

```python
# print(data["company_location"].value_counts().head(20))
# print(data["employee_residence"].value_counts().head(20))
# print(data["remote_ratio"].value_counts().head(20))
```

```python
# Keep only the 7 countries with the most data
mask = data["company_location"].isin(["FR", "US", "GB", "CA", "ES", "IN", "DE"])
data = data[mask]

print(data["company_location"].value_counts().head(20))
print(data["employee_residence"].value_counts().head(20))

mask = data["employee_residence"].isin(["FR", "US", "GB", "CA", "ES", "IN", "DE"])
data = data[mask]

print(data["company_location"].value_counts().head(20))
print(data["employee_residence"].value_counts().head(20))

print(data.describe())
print(data.info())
```

```python
# Transform FR to France, ...
data["company_location"] = data["company_location"].map(countries_dict)
data["employee_residence"] = data["employee_residence"].map(countries_dict)

print(data["company_location"].value_counts().head(20))
print(data["employee_residence"].value_counts().head(20))
```

```python
# data.to_csv("ds_salaries_filtered.csv")
```

```python
# Explore
myval_count(data)
```

```python
# data[data.duplicated(data.columns, keep=False)]
display(data[data.duplicated() == True])

display(data[(data["company_location"] == "Germany") & (data["salary_in_usd"] == 90734)])
```

```python
# drop 1/3 of the data as they are duplicates
data.drop_duplicates(inplace=True)
print(data.info())
myval_count(data)

display(data[(data["company_location"] == "Germany") & (data["salary_in_usd"] == 90734)])
```

```python

print(data['job_title'].nunique())
print(data['job_title'].unique())
print(data['job_title'].value_counts().head(30))

# Should regroup titles like eg "ML Engineer" and "Machine LEarning Engineer" but no time
```

```python

data['experience_level'] = data['experience_level'].replace({'SE':'Senior', 
                                                             'MI':'Mid-level', 
                                                             'EN':'Entry-level',
                                                             'EX':'Executive-level'})
data['employment_type'] = data['employment_type'].replace({'FT':'Full-time', 
                                                           'CT':'Contractual',
                                                           'PT':'Part-time', 
                                                           'FL':'Freelancer'})
data['remote_ratio'] = data['remote_ratio'].replace({0:'On-site', 
                                                     100:'Remote', 
                                                     50:'Hybrid'})
```
