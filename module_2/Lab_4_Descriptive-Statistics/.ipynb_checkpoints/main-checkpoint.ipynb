## Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
def roll_dice(n):
    """Simulate rolling a dice n times."""
    results = choices(range(1, 7), k=n)  # Simulate rolling a 6-sided dice n times
    return pd.DataFrame(results, columns=["Roll"])

# Test the function
df = roll_dice(10)
print(df)
```

#### 2.- Plot the results sorted by value.

```python
def plot_sorted_rolls(df):
    """Plot sorted dice roll results."""
    sorted_rolls = df.sort_values(by="Roll").reset_index(drop=True)
    plt.figure(figsize=(10, 5))
    sorted_rolls.plot(kind="bar")
    plt.ylabel("Dice Value")
    plt.xlabel("Roll Number")
    plt.title("Sorted Dice Rolls")
    plt.xticks(rotation=0)
    plt.show()

plot_sorted_rolls(df)
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python
def plot_frequency_distribution(df):
    """Plot frequency distribution of dice roll results."""
    freq = df["Roll"].value_counts().sort_index()
    plt.figure(figsize=(10, 5))
    freq.plot(kind="bar")
    plt.ylabel("Frequency")
    plt.xlabel("Dice Value")
    plt.title("Frequency Distribution of Dice Rolls")
    plt.xticks(rotation=0)
    plt.show()

plot_frequency_distribution(df)
```

```python
"""
the first plot is sorted dice roll results and the second plot give us de frequency of the dice value like for exemple: 
the value 2 happened 3 times.
"""
```

## Challenge 2
Now, using the dice results obtained in *challenge 1*, your are going to define some functions that will help you calculate the mean of your data in two different ways, the median and the four quartiles. 

#### 1.- Define a function that computes the mean by summing all the observations and dividing by the total number of observations. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
def compute_mean(data):
    """Compute the mean of a list of numbers."""
    return sum(data) / len(data)

dice_mean = compute_mean(df["Roll"].tolist())
print("Mean:", dice_mean)
```

#### 2.- First, calculate the frequency distribution. Then, calculate the mean using the values of the frequency distribution you've just computed. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
def compute_mean_from_freq(freq_dist):
    """Compute the mean from a frequency distribution."""
    total_sum = sum([value * freq for value, freq in freq_dist.items()])
    total_numbers = sum(freq_dist.values())
    return total_sum / total_numbers

# Calculate the frequency distribution
freq_dist = df["Roll"].value_counts().to_dict()
print(freq_dist)

dice_mean_from_freq = compute_mean_from_freq(freq_dist)
print("Mean from Frequency Distribution:", dice_mean_from_freq)
```

#### 3.- Define a function to calculate the median. You are not allowed to use any methods or functions that directly calculate the median value. 
**Hint**: you might need to define two computation cases depending on the number of observations used to calculate the median.

```python
def compute_median(data):
    """Compute the median of a list of numbers."""
    sorted_data = sorted(data)
    n = len(sorted_data)
    
    # If even number of data points
    if n % 2 == 0:
        median1 = sorted_data[n//2]
        median2 = sorted_data[n//2 - 1]
        median = (median1 + median2) / 2
    # If odd number of data points
    else:
        median = sorted_data[n//2]
    
    return median

dice_median = compute_median(df["Roll"].tolist())
print("Median:", dice_median)
```

#### 4.- Define a function to calculate the four quartiles. You can use the function you defined above to compute the median but you are not allowed to use any methods or functions that directly calculate the quartiles. 

```python
def compute_quartiles(data):
    """Compute the quartiles of a list of numbers."""
    sorted_data = sorted(data)
    
    # Compute the median
    Q2 = compute_median(sorted_data)
    
    # Split data into lower and upper halves
    if len(sorted_data) % 2 == 0:
        lower_half = sorted_data[:len(sorted_data)//2]
        upper_half = sorted_data[len(sorted_data)//2:]
    else:
        lower_half = sorted_data[:len(sorted_data)//2]
        upper_half = sorted_data[len(sorted_data)//2+1:]
    
    # Compute Q1 and Q3
    Q1 = compute_median(lower_half)
    Q3 = compute_median(upper_half)
    
    return Q1, Q2, Q3

Q1, Q2, Q3 = compute_quartiles(df["Roll"].tolist())
print("Q1:", Q1)
print("Q2:", Q2)
print("Q3:", Q3)
```
